# Customer Segmentation for an E-Commerce Business

## Data collection

```python
import pandas as pd
```

```python
# Load the data
df = pd.read_csv("data.csv", encoding="ISO-8859-1")
```

```python
df
```

## Data Cleaning

```python
# Drop rows with missing values
df = df.dropna()
```

```python
df
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
df
```

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df
```

```python
df.loc[:, 'InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
df.loc[:, 'TotalPrice'] = df['Quantity'] * df['UnitPrice']
df
```

## Feature Engineering (RFM Analysis)

```python
from datetime import datetime

# Define a Reference Date
reference_date = datetime(2011, 12, 9)
```

```python
#we need to use max date
max_date = df['InvoiceDate'].max()
print("Maximum Date:", max_date)
```

```python
#calculate RFM Features:
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalPrice': 'sum'
}).rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency', 'TotalPrice': 'Monetary'})
```

```python
rfm
```

## Clustering

```python
import sklearn as sl
```

```python
#Normalize the RFM Data:
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
```

```python
rfm_scaled
```

```python
#Apply K-Means Clustering:
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=11, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
print(rfm['Cluster'].value_counts())
```

## Visualization

```python
#pair plot
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(rfm, hue='Cluster', palette='Set2')
plt.show()
```

```python
#Cluster Analysis:
rfm.groupby('Cluster').mean().plot(kind='bar', figsize=(10, 6))
plt.title('Average RFM Values by Cluster')
plt.show()
```
